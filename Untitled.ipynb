```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('amz_uk_price_prediction_dataset.csv')
```

```python
df.head(10)
```

```python
crosstab = pd.crosstab(df['category'], df['isBestSeller'])
best_seller_proportion = (crosstab[True] / crosstab.sum(axis=1)).sort_values(ascending=False)
print(best_seller_proportion.head(10))
```

```python
chi2, p, dof, expected = chi2_contingency(crosstab)
print(f"Chi2 Statistic: {chi2}, p-value: {p}")
```

```python
# Cramér's V calculation
n = df.shape[0]
cramers_v = np.sqrt(chi2 / (n * (min(crosstab.shape) - 1)))
print(f"Cramér's V: {cramers_v}")
```

```python
# Select top 10 categories by total count (sum of all best-seller statuses)
top_categories = crosstab.sum(axis=1).nlargest(10).index

# Filter the crosstab for these top categories
crosstab_filtered = crosstab.loc[top_categories]

# Normalize the filtered crosstab
crosstab_normalized_filtered = crosstab_filtered.div(crosstab_filtered.sum(axis=1), axis=0)

# Plot the normalized filtered crosstab
crosstab_normalized_filtered.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title("Proportion of Best-Seller Status for Top Categories")
plt.ylabel("Proportion")
plt.xlabel("Category")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.legend(title="Best-Seller Status", loc='upper right')
plt.tight_layout()  # Adjust layout to fit everything nicely
plt.show()
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_no_outliers = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
```

```python
# Top 20 categories based on count
top_20_categories = df_no_outliers['category'].value_counts().head(20).index

# Violin plot for price across categories
sns.violinplot(data=df_no_outliers[df_no_outliers['category'].isin(top_20_categories)],
               x='category', y='price', scale='width')
plt.xticks(rotation=45)
plt.title("Price Distribution Across Top 20 Categories")
plt.show()

# Median price for all categories
category_median_prices = df_no_outliers.groupby('category')['price'].median().sort_values(ascending=False)
print(category_median_prices.head())
```

```python
# Top 10 categories based on count
top_10_categories = df_no_outliers['category'].value_counts().head(10).index

# Average price for top 10 categories
avg_price_top_10 = df_no_outliers[df_no_outliers['category'].isin(top_10_categories)].groupby('category')['price'].mean()
avg_price_top_10.plot(kind='bar', figsize=(10, 6))
plt.title("Average Price for Top 10 Categories")
plt.ylabel("Average Price")
plt.show()

# Find the category with the highest average price
highest_avg_price_category = avg_price_top_10.idxmax()
print(f"Category with the highest average price: {highest_avg_price_category}")
```

```python
# Box plot for ratings by category
sns.boxplot(data=df_no_outliers[df_no_outliers['category'].isin(top_10_categories)],
            x='category', y='stars')
plt.xticks(rotation=45)
plt.title("Ratings Distribution Across Top 10 Categories")
plt.show()

# Find the category with the highest median rating
median_ratings = df_no_outliers.groupby('category')['stars'].median().sort_values(ascending=False)
print(median_ratings.head())
```

```python
# Correlation between price and stars
correlation = df_no_outliers['price'].corr(df_no_outliers['stars'])
print(f"Correlation between price and stars: {correlation}")
```

```python
# Scatter plot
sns.scatterplot(data=df_no_outliers, x='stars', y='price')
plt.title("Relationship Between Price and Ratings")
plt.show()

# Correlation heatmap
sns.heatmap(df_no_outliers[['price', 'stars']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# QQ plot for price distribution
from statsmodels.graphics.gofplots import qqplot
qqplot(df_no_outliers['price'], line='s')
plt.title("QQ Plot of Prices")
plt.show()
```
